```julia
vertices, faces = isosurface(Float32.(mask), MarchingCubes(iso=0.5))
```

```julia
function mesh_volume(vertices, faces)
    volume = 0.0
    for face in faces
        a, b, c = vertices[face[1]], vertices[face[2]], vertices[face[3]]
        volume += dot(a, cross(SVector(b), SVector(c))) / 6
    end
    return abs(volume)
end
```

```julia
v=mesh_volume(vertices, faces)
```
